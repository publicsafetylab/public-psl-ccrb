```python
def complaints_officers_crimes_per_year(all_complaints, substantiated_complaints):
    all_complaints = all_complaints[(all_complaints["Year"] > 1985) & (all_complaints["Year"] < 2020)]
    all_complaints["Num_NYPD_Officers_Year"] = np.where(all_complaints["Year"]==2003, 36700, all_complaints["Num_NYPD_Officers_Year"])
    substantiated_complaints = substantiated_complaints[(substantiated_complaints["Year"] > 1985) & (substantiated_complaints["Year"] < 2020)]
    substantiated_complaints["Num_NYPD_Officers_Year"] = np.where(substantiated_complaints["Year"]==2003, 36700, substantiated_complaints["Num_NYPD_Officers_Year"])
    all_group = all_complaints.groupby("Year")["Unique Id"].count().reset_index().rename(columns={"Unique Id": "All"})
    substantiated_group = substantiated_complaints.groupby("Year")["Unique Id"].count().reset_index().rename(columns={"Unique Id": "Substantiated"})
    complaints_per_year = pd.merge(all_group, substantiated_group, on="Year")
    officers_group = all_complaints.drop_duplicates(["Year", "Num_NYPD_Officers_Year"])[["Year", "Num_NYPD_Officers_Year"]].sort_values(by="Year")
    crimes_group = all_complaints.drop_duplicates(["Year", "Num_Offenses_Year"])[["Year", "Num_Offenses_Year"]].sort_values(by="Year")
    officers_crimes_per_year = pd.merge(officers_group, crimes_group, on="Year")
    all_per_year = pd.merge(complaints_per_year, officers_crimes_per_year, on="Year")
    fig = make_subplots(rows=2, cols=2)
    fig.add_trace(
        go.Scatter(x=all_per_year.Year, y=all_per_year.All, name="All Complaints"),
        row=1, col=1
    )
    fig.add_trace(
        go.Scatter(x=all_per_year.Year, y=all_per_year.Substantiated, name="Substantiated"),
        row=1, col=2
    )
    fig.add_trace(
        go.Scatter(x=all_per_year.Year, y=all_per_year.Num_NYPD_Officers_Year, name="NYPD Officers"),
        row=2, col=1
    )
    fig.add_trace(
        go.Scatter(x=all_per_year.Year, y=all_per_year.Num_Offenses_Year, name="Crime Reports"),
        row=2, col=2
    )
    fig.update_layout(
        title={
            'text': "<b>Figure 2</b>: Numbers of Misconduct Complaints, Substantiated Misconduct Complaints, Sworn NYPD Officers & Reported Crimes (1986-2019)",
            'y':0.9,
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'})
    fig.update_xaxes(title_text="<span style='font-size: 12px;'>Year</span>")
    fig.update_yaxes(title_text="<span style='font-size: 12px;'>All Misconduct Complaints</span>", row=1, col=1)
    fig.update_yaxes(title_text="<span style='font-size: 12px;'>Substantiated Complaints</span>", row=1, col=2)
    fig.update_yaxes(title_text="<span style='font-size: 12px;'>Sworn NYPD Officers</span>", row=2, col=1)
    fig.update_yaxes(title_text="<span style='font-size: 12px;'>Reported Crimes</span>", row=2, col=2)
    fig.update(layout_showlegend=False)
    fig.write_html(f"viz/fig-2.html")
    fig.show()
```

```python
def complaints_crimes_per_year_ols(all_complaints):
    all_complaints = all_complaints[(all_complaints["Year"] > 2005) & (all_complaints["Year"] < 2019)]
    all_complaints = all_complaints[(all_complaints["Precinct"] != "-1") & (all_complaints["Precinct"] != "121")]
    all_group = all_complaints.groupby(["Year", "Precinct"])["Unique Id"].count().reset_index().rename(columns={"Unique Id": "All Complaints"})
    all_group = all_group.groupby("Precinct")["All Complaints"].mean().reset_index()
    cols = [c for c in all_complaints.columns if "Num_Crime_Complaints_" in c and "_Precinct_Year" in c]
    pcts_dfs = []
    for c in cols:
        pcts_dfs.append(all_complaints.drop_duplicates(["Year", "Precinct", c])[["Year", "Precinct", c]].sort_values(by="Year"))
    pcts_df = pd.concat(pcts_dfs, axis=1)
    pcts_df = pcts_df.loc[:,~pcts_df.columns.duplicated()]
    pcts_df["Crime Reports"] = pcts_df[cols].sum(axis=1)
    pcts_group = pcts_df[["Year", "Precinct", "Crime Reports"]].groupby("Precinct")["Crime Reports"].mean().reset_index()
    all_per_year = pd.merge(all_group, pcts_group, on="Precinct").rename(columns={"All Complaints": "Mean Annual Misconduct Complaints", "Crime Reports": "Mean Annual Reported Crimes"}).set_index("Precinct")
    rg = sns.regplot(x=all_per_year["Mean Annual Reported Crimes"], y=all_per_year["Mean Annual Misconduct Complaints"])
    X = rg.get_lines()[0].get_xdata()
    Y = rg.get_lines()[0].get_ydata()
    P = rg.get_children()[1].get_paths()
    p_codes={1:'M', 2: 'L', 79: 'Z'}
    path=''
    for s in P[0].iter_segments():
        c=p_codes[s[1]]
        xx, yy=s[0]
        path+=c+str('{:.5f}'.format(xx))+' '+str('{:.5f}'.format(yy))
    shapes=[dict(type='path',
                 path=path,
                 line=dict(width=0.1,color='rgba(68, 122, 219, 0.25)' ),
                 fillcolor='rgba(68, 122, 219, 0.25)')]   
    fig = px.scatter(all_per_year, x=all_per_year["Mean Annual Reported Crimes"], y=all_per_year["Mean Annual Misconduct Complaints"], text=all_per_year.index, trendline="ols")
    fig.update_traces(textposition='top center', textfont_size=6)
    fig.update_layout(shapes=shapes)
    fig.update_xaxes(title_text="<span style='font-size: 12px;'>Mean Annual Number of Reported Crimes</span>")
    fig.update_yaxes(title_text="<span style='font-size: 12px;'>Mean Annual Number of Misconduct Complaints</span>")
    fig.update_layout(
        title={
            'text': "<b>Figure 4</b>: Per-Precinct Mean Annual Misconduct Complaints vs. Mean Annual Reported Crimes (2006-2019)",
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'})
    fig.write_html(f"viz/fig-4.html")
    fig.show()
    results = px.get_trendline_results(fig)
    const, x1 = results.px_fit_results.iloc[0].params
    return const, x1, results.px_fit_results.iloc[0].summary()
```

```python
def excess_complaints_demo_res_per_year_ols(all_complaints, const, x1, demo, figno):
    if demo.startswith("NH"):
        demo_alt = demo.replace("NH_", "Non-Hispanic ")
    else:
        demo_alt = demo
    all_complaints = all_complaints[(all_complaints["Year"] > 2005) & (all_complaints["Year"] < 2020)]
    all_complaints = all_complaints[(all_complaints["Precinct"] != "-1") & (all_complaints["Precinct"] != "121")]
    all_group = all_complaints.groupby(["Year", "Precinct"])["Unique Id"].count().reset_index().rename(columns={"Unique Id": "All Complaints"})
    all_group = all_group.groupby("Precinct")["All Complaints"].mean().reset_index()
    cols = [c for c in all_complaints.columns if "Num_Crime_Complaints_" in c and "_Precinct_Year" in c]
    pcts_dfs = []
    for c in cols:
        pcts_dfs.append(all_complaints.drop_duplicates(["Year", "Precinct", c])[["Year", "Precinct", c]].sort_values(by="Year"))
    pcts_df = pd.concat(pcts_dfs, axis=1)
    pcts_df = pcts_df.loc[:,~pcts_df.columns.duplicated()]
    pcts_df["Crime Reports"] = pcts_df[cols].sum(axis=1)
    pcts_group = pcts_df[["Year", "Precinct", "Crime Reports"]].groupby("Precinct")["Crime Reports"].mean().reset_index()
    all_per_year = pd.merge(all_group, pcts_group, on="Precinct").rename(columns={"All Complaints": "Mean Annual Misconduct Complaints", "Crime Reports": "Mean Annual Reported Crimes"})
    all_per_year["y-hat"] = const + x1*all_per_year["Mean Annual Reported Crimes"] 
    all_per_year["resid"] = all_per_year["Mean Annual Misconduct Complaints"] - all_per_year["y-hat"]
    pct_black = all_complaints.drop_duplicates(["Precinct", demo])[["Precinct", demo]].rename(columns={demo: f"Percent {demo.split('_')[0]} Residents"}).sort_values(by="Precinct")
    all_per_year = pd.merge(all_per_year, pct_black, on="Precinct")
    all_per_year = all_per_year[["Precinct", "resid", f"Percent {demo.split('_')[0]} Residents"]].rename(columns={"resid": "Mean 'Excess' Annual Misconduct Complaints"})
    all_per_year = all_per_year[all_per_year[f"Percent {demo.split('_')[0]} Residents"].notnull()]
    all_per_year["Precinct"] = all_per_year["Precinct"].astype(int)
    all_per_year = all_per_year.sort_values(by="Precinct")
    all_per_year[f"Percent {demo.split('_')[0]} Residents"] = 100*all_per_year[f"Percent {demo.split('_')[0]} Residents"]    
    all_per_year = all_per_year.rename(columns={f"Percent {demo.split('_')[0]} Residents": f"Percent {demo_alt.split('_')[0]} Residents"})
    rg = sns.regplot(x=all_per_year[f"Percent {demo_alt.split('_')[0]} Residents"], y=all_per_year["Mean 'Excess' Annual Misconduct Complaints"])
    X = rg.get_lines()[0].get_xdata()
    Y = rg.get_lines()[0].get_ydata()
    P = rg.get_children()[1].get_paths()
    p_codes={1:'M', 2: 'L', 79: 'Z'}
    path=''
    for s in P[0].iter_segments():
        c=p_codes[s[1]]
        xx, yy=s[0]
        path+=c+str('{:.5f}'.format(xx))+' '+str('{:.5f}'.format(yy))
    shapes=[dict(type='path',
                 path=path,
                 line=dict(width=0.1,color='rgba(68, 122, 219, 0.25)' ),
                 fillcolor='rgba(68, 122, 219, 0.25)')] 
    fig = px.scatter(all_per_year, x=all_per_year[f"Percent {demo_alt.split('_')[0]} Residents"], y=all_per_year["Mean 'Excess' Annual Misconduct Complaints"], color = all_per_year.Precinct, text=all_per_year.Precinct, trendline="ols")
    fig.update_traces(textposition='top center', textfont_size=6)
    fig.update_layout(shapes=shapes)
    fig.update_xaxes(title_text=f"<span style='font-size: 12px;'>Percent {demo_alt.split('_')[0]} Residents</span>")
    fig.update_yaxes(title_text="<span style='font-size: 12px;'>Mean 'Excess' Annual Number of Misconduct Complaints</span>")
    fig.update_layout(
        title={
            'text': f"<b>Figure {figno.upper()}</b>: Per-Precinct Mean Annual 'Excess' Misconduct Complaints (2006-2019) vs. Percent {demo_alt.split('_')[0]} Residents (2010 Census)",
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'})
    fig.write_html(f"viz/fig-{figno}.html")
    fig.show()
    results = px.get_trendline_results(fig)
    return results.px_fit_results.iloc[0].summary()
```

```python
def substantiated_excess_complaints_demo_res_per_year_ols(substantiated_complaints, const, x1, demo, figno):
    if demo.startswith("NH"):
        demo_alt = demo.replace("NH_", "Non-Hispanic ")
    else:
        demo_alt = demo
    substantiated_complaints = substantiated_complaints[(substantiated_complaints["Year"] > 2005) & (substantiated_complaints["Year"] < 2020)]
    substantiated_complaints = substantiated_complaints[(substantiated_complaints["Precinct"] != "-1") & (substantiated_complaints["Precinct"] != "121")]
    all_group = substantiated_complaints.groupby(["Year", "Precinct"])["Unique Id"].count().reset_index().rename(columns={"Unique Id": "All Complaints"})
    all_group = all_group.groupby("Precinct")["All Complaints"].mean().reset_index()
    cols = [c for c in substantiated_complaints.columns if "Num_Crime_Complaints_" in c and "_Precinct_Year" in c]
    pcts_dfs = []
    for c in cols:
        pcts_dfs.append(substantiated_complaints.drop_duplicates(["Year", "Precinct", c])[["Year", "Precinct", c]].sort_values(by="Year"))
    pcts_df = pd.concat(pcts_dfs, axis=1)
    pcts_df = pcts_df.loc[:,~pcts_df.columns.duplicated()]
    pcts_df["Crime Reports"] = pcts_df[cols].sum(axis=1)
    pcts_group = pcts_df[["Year", "Precinct", "Crime Reports"]].groupby("Precinct")["Crime Reports"].mean().reset_index()
    all_per_year = pd.merge(all_group, pcts_group, on="Precinct").rename(columns={"All Complaints": "Mean Annual Misconduct Complaints", "Crime Reports": "Mean Annual Reported Crimes"})
    all_per_year["y-hat"] = const + x1*all_per_year["Mean Annual Reported Crimes"] 
    all_per_year["resid"] = all_per_year["Mean Annual Misconduct Complaints"] - all_per_year["y-hat"]
    pct_black = all_complaints.drop_duplicates(["Precinct", demo])[["Precinct", demo]].rename(columns={demo: f"Percent {demo.split('_')[0]} Residents"}).sort_values(by="Precinct")
    all_per_year = pd.merge(all_per_year, pct_black, on="Precinct")
    all_per_year = all_per_year[["Precinct", "resid", f"Percent {demo.split('_')[0]} Residents"]].rename(columns={"resid": "Mean 'Excess' Annual Substantiated Misconduct Complaints"})
    all_per_year = all_per_year[all_per_year[f"Percent {demo.split('_')[0]} Residents"].notnull()]
    all_per_year["Precinct"] = all_per_year["Precinct"].astype(int)
    all_per_year = all_per_year.sort_values(by="Precinct")
    all_per_year[f"Percent {demo.split('_')[0]} Residents"] = 100*all_per_year[f"Percent {demo.split('_')[0]} Residents"]    
    all_per_year = all_per_year.rename(columns={f"Percent {demo.split('_')[0]} Residents": f"Percent {demo_alt.split('_')[0]} Residents"})
    rg = sns.regplot(x=all_per_year[f"Percent {demo_alt.split('_')[0]} Residents"], y=all_per_year["Mean 'Excess' Annual Substantiated Misconduct Complaints"])
    X = rg.get_lines()[0].get_xdata()
    Y = rg.get_lines()[0].get_ydata()
    P = rg.get_children()[1].get_paths()
    p_codes={1:'M', 2: 'L', 79: 'Z'}
    path=''
    for s in P[0].iter_segments():
        c=p_codes[s[1]]
        xx, yy=s[0]
        path+=c+str('{:.5f}'.format(xx))+' '+str('{:.5f}'.format(yy))
    shapes=[dict(type='path',
                 path=path,
                 line=dict(width=0.1,color='rgba(68, 122, 219, 0.25)' ),
                 fillcolor='rgba(68, 122, 219, 0.25)')] 
    fig = px.scatter(all_per_year, x=all_per_year[f"Percent {demo_alt.split('_')[0]} Residents"], y=all_per_year["Mean 'Excess' Annual Substantiated Misconduct Complaints"], color = all_per_year.Precinct, text=all_per_year.Precinct, trendline="ols")
    fig.update_traces(textposition='top center', textfont_size=6)
    fig.update_layout(shapes=shapes)
    fig.update_xaxes(title_text=f"<span style='font-size: 12px;'>Percent {demo_alt.split('_')[0]} Residents</span>")
    fig.update_yaxes(title_text="<span style='font-size: 12px;'>Mean 'Excess' Annual Number of Substantiated Misconduct Complaints</span>")
    fig.update_layout(
        title={
            'text': f"<b>Figure {figno.upper()}</b>: Per-Precinct Mean Annual 'Excess' Substantiated Misconduct Complaints (2006-2019) vs. Percent {demo_alt.split('_')[0]} Residents (2010 Census)",
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'})
    fig.write_html(f"viz/fig-{figno}.html")
    fig.show()
    results = px.get_trendline_results(fig)
    return results.px_fit_results.iloc[0].summary()
```

```python
def complaints_officer_concentration_per_year_ols(all_complaints, const, x1):
    all_complaints = all_complaints[(all_complaints["Year"] > 2005) & (all_complaints["Year"] < 2020)]
    all_complaints = all_complaints[(all_complaints["Precinct"] != "-1") & (all_complaints["Precinct"] != "121")]
    all_group = all_complaints.groupby(["Year", "Precinct"])["Unique Id"].count().reset_index().rename(columns={"Unique Id": "All Complaints"})
    all_group = all_group.groupby("Precinct")["All Complaints"].mean().reset_index()
    cols = [c for c in all_complaints.columns if "Num_Crime_Complaints_" in c and "_Precinct_Year" in c]
    pcts_dfs = []
    for c in cols:
        pcts_dfs.append(all_complaints.drop_duplicates(["Year", "Precinct", c])[["Year", "Precinct", c]].sort_values(by="Year"))
    pcts_df = pd.concat(pcts_dfs, axis=1)
    pcts_df = pcts_df.loc[:,~pcts_df.columns.duplicated()]
    pcts_df["Crime Reports"] = pcts_df[cols].sum(axis=1)
    pcts_group = pcts_df[["Year", "Precinct", "Crime Reports"]].groupby("Precinct")["Crime Reports"].mean().reset_index()
    pct_group = all_complaints.groupby(["Precinct", "Unique Id"])["Unique Id"].count().rename("Count").reset_index()
    pct_group = pct_group.groupby("Precinct")["Count"].mean().reset_index()
    all_per_year = pd.merge(all_group, pcts_group, on="Precinct")
    all_per_year["y-hat"] = const + x1*all_per_year["Crime Reports"] 
    all_per_year["resid"] = all_per_year["All Complaints"] - all_per_year["y-hat"]
    all_per_year = pd.merge(all_per_year, pct_group, on="Precinct")
    all_per_year = all_per_year[["Precinct", "resid", "Count"]]
    all_per_year = all_per_year.rename(columns={"resid": "Mean 'Excess' Annual Misconduct Complaints", "Count": "Mean Misconduct Complaints Per Accused Officer"})
    rg = sns.regplot(x=all_per_year["Mean Misconduct Complaints Per Accused Officer"], y=all_per_year["Mean 'Excess' Annual Misconduct Complaints"])
    X = rg.get_lines()[0].get_xdata()
    Y = rg.get_lines()[0].get_ydata()
    P = rg.get_children()[1].get_paths()
    p_codes={1:'M', 2: 'L', 79: 'Z'}
    path=''
    for s in P[0].iter_segments():
        c=p_codes[s[1]]
        xx, yy=s[0]
        path+=c+str('{:.5f}'.format(xx))+' '+str('{:.5f}'.format(yy))
    shapes=[dict(type='path',
                 path=path,
                 line=dict(width=0.1,color='rgba(68, 122, 219, 0.25)' ),
                 fillcolor='rgba(68, 122, 219, 0.25)')]   
    fig = px.scatter(all_per_year, x=all_per_year["Mean Misconduct Complaints Per Accused Officer"], y=all_per_year["Mean 'Excess' Annual Misconduct Complaints"], text=all_per_year.Precinct, trendline="ols")
    fig.update_traces(textposition='top center', textfont_size=6)
    fig.update_layout(shapes=shapes)
    fig.update_xaxes(title_text="<span style='font-size: 12px;'>Mean Number of Misconduct Complaints Per Accused Officer</span>")
    fig.update_yaxes(title_text="<span style='font-size: 12px;'>Mean 'Excess' Annual Number of Misconduct Complaints</span>")
    fig.update_layout(
        title={
            'text': "<b>Figure 6</b>: Per-Precinct Mean Annual 'Excess' Misconduct Complaints vs. Mean Misconduct Complaints Per Accused Officer (2006-2019)",
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'})
    fig.write_html(f"viz/fig-6.html")
    fig.show()
    results = px.get_trendline_results(fig)
    return results.px_fit_results.iloc[0].summary()
```
